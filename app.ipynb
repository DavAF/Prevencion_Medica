```python
import streamlit as st

st.set_page_config(page_title="Chatbot Renovaci√≥n Carnet", page_icon="üöó")

st.title("ü§ñ Chatbot: Certificado M√©dico para Carnet de Conducir")

# Simulaci√≥n de datos por comunidad
centros = {
    "Madrid": ["Centro M√©dico Goya", "Cl√≠nica Reconduce", "Centro Sanitas Cuatro Caminos"],
    "Andaluc√≠a": ["Cl√≠nica Andaluc√≠a Conduce", "Centro M√©dico Sur", "M√©dicos Aljarafe"],
    "Catalu√±a": ["M√©dico Carnet BCN", "Cl√≠nica Diagonal Drive", "Centro M√©dico Gracia"]
}

# Estado de conversaci√≥n
if "step" not in st.session_state:
    st.session_state.step = 1

def reiniciar_chat():
    st.session_state.step = 1
    st.session_state.comunidad = ""
    st.session_state.nombre = ""

# Paso 1: nombre del usuario
if st.session_state.step == 1:
    st.write("üëã ¬°Hola! Soy tu asistente para renovar el carnet de conducir.")
    st.session_state.nombre = st.text_input("¬øC√≥mo te llamas?")
    if st.session_state.nombre:
        st.session_state.step = 2
        st.experimental_rerun()

# Paso 2: comunidad aut√≥noma
elif st.session_state.step == 2:
    st.write(f"Encantado, {st.session_state.nombre}.")
    st.session_state.comunidad = st.selectbox("¬øEn qu√© comunidad aut√≥noma est√°s?", list(centros.keys()))
    if st.session_state.comunidad:
        st.session_state.step = 3
        st.experimental_rerun()

# Paso 3: Mostrar informaci√≥n relevante
elif st.session_state.step == 3:
    st.write(f"üîç En {st.session_state.comunidad}, estos son los pasos para renovar tu carnet:")

    st.markdown("""
    1. **Solicita un certificado m√©dico** en un centro autorizado.
    2. Lleva tu DNI/NIE y el carnet de conducir.
    3. Hazte una revisi√≥n (vista, reflejos, presi√≥n).
    4. Te entregar√°n el certificado directamente o lo enviar√°n a la DGT.
    5. **Precio estimado**: entre 30‚ÄØ‚Ç¨ y 60‚ÄØ‚Ç¨.
    """)

    st.write("üè• Centros m√©dicos recomendados:")
    for c in centros[st.session_state.comunidad]:
        st.write(f"- {c}")

    if st.button("Finalizar"):
        st.session_state.step = 4
        st.experimental_rerun()

# Paso 4: Final
elif st.session_state.step == 4:
    st.success("‚úÖ ¬°Listo! Ya sabes c√≥mo renovar tu carnet.")
    st.markdown("¬øQuieres volver a empezar?")
    if st.button("üîÑ Reiniciar"):
        reiniciar_chat()

```
